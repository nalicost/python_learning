```python
from langchain.prompts import StringPromptTemplate

PROMPT = """\
name: {name}
age: {age}
"""


class My(StringPromptTemplate):
    
    def format(self, **kwargs):
        return PROMPT.format(name=kwargs['name'], age=kwargs['age'])


m = My(input_variables=['name', 'age'])
p = m.format(name='lucy', age=10)
print(p)
```
