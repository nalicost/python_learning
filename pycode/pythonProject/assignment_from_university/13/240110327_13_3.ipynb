```python
from sklearn.datasets import load_iris 
from sklearn.preprocessing import StandardScaler 
from sklearn.cluster import KMeans, AgglomerativeClustering 
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
# 
# Iris 
iris = load_iris()
X = iris.data[:,:2] # 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
#
def plot_clusters(X, y_pred, title):
    plt.scatter (X[:,0], X[:,1], c=y_pred, cmap='viridis', marker='o')
    plt.title(title)
    plt.xlabel ('Sepal Length (standardized)') 
    plt.ylabel('Sepal Width (standardized)')
    plt.show()                              
# K-Means 
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)
plot_clusters(X_scaled, y_kmeans, "K-Means Clustering")
# 
agg_clustering = AgglomerativeClustering(n_clusters=3)
y_agg = agg_clustering.fit_predict(X_scaled)
plot_clusters(X_scaled, y_agg, "Hierarchical Clustering")
# DBSCAN 
dbscan = DBSCAN(eps=0.5, min_samples =5)
y_dbscan = dbscan.fit_predict(X_scaled)
plot_clusters(X_scaled, y_dbscan, "DBSCAN Clustering")
```
